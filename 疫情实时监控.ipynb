```python
import time
# 历史数据和当日详细数据
def getTencentData():
    url = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5'
    url_other = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_other'
    header = {
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"
    }

    r = requests.get(url,headers=header)
    res = json.loads(r.text)  #json字符转字典
    data_all = json.loads(res["data"])

    r_other = requests.get(url_other,headers=header)
    res_other = json.loads(r_other.text)  #json字符转字典
    data_all_other = json.loads(res_other["data"])

    history = {}   #历史数据
    for i in data_all_other["chinaDayList"]:
        ds = "2020."+i["date"]   #date格式：04.22
        t = time.strptime(ds, "%Y.%m.%d")
        ds = time.strftime("%Y-%m-%d",t)  #改变时间格式 2020-01-13 ，以便插入数据库
        #print(ds)
        confirm = i["confirm"]   #确诊
        suspect = i["suspect"]   #疑似
        heal = i["heal"]   #治愈
        dead = i["dead"]   #死亡
        history[ds] = {"confirm":confirm, "suspect":suspect, "heal":heal, "dead":dead}
        #print(history)

    # 历史新增数据
    for i in data_all_other["chinaDayAddList"]:
        ds = "2020."+i["date"]   #date格式：04.22
        t = time.strptime(ds, "%Y.%m.%d")
        ds = time.strftime("%Y-%m-%d",t)  #改变时间格式 2020-01-13 ，以便插入数据库
        #print(ds)
        confirm_add = i["confirm"]   #确诊
        suspect_add = i["suspect"]   #疑似
        heal_add = i["heal"]   #治愈
        dead_add = i["dead"]   #死亡
        history[ds].update({"confirm_add":confirm_add, "suspect_add":suspect_add, "heal_add":heal_add, "dead_add":dead_add})
        #print(history[ds])

    details = []   #当日详细数据
    update_time = data_all["lastUpdateTime"]
    data_country = data_all["areaTree"]   # list 国家
    data_province = data_country[0]["children"]   #中国各省
    for pro_infos in data_province:
        province = pro_infos["name"]
        #print(pro_infos["name"])
        for city_infos in pro_infos["children"]:
            #print(city_infos["name"])
            #print(city_infos["total"])
            #print(city_infos["today"])

            citey = city_infos["name"]
            confirm = city_infos["total"]["confirm"]
            confirm_add = city_infos["today"]["confirm"]
            heal = city_infos["total"]["heal"]
            dead = city_infos["total"]["dead"]
            details.append([update_time,province,citey,confirm,confirm_add,heal,dead])
    #print(details)
    return history,details

```

```python
import pymysql
def getConn():
    # 建立连接
    conn = pymysql.connect(host="127.0.0.1",
                           user = "root",
                           password = "",
                           db = "cov2019")
    # 创建游标
    cursor = conn.cursor()
    return conn,cursor

def closeConn(conn, cursor):
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```

```python
# 更新details表
def updateDetails():
    cursor = None
    conn = None
    try:
        li = getTencentData()[1]    # 0 是历史数据字典， 1 是最新详细数据字典
        conn, cursor = getConn()
        sql = "insert into details(update_time,province,city,confirm,confirm_add,heal,dead) values(%s,%s,%s,%s,%s,%s,%s)"
        sql_query = "select %s=(select update_time from details order by id desc limit 1)"
        cursor.execute(sql_query, li[0][0])
        if not cursor.fetchone()[0]:
            print(f"{time.asctime()}开始更新最新数据")
            for item in li:
                cursor.execute(sql,item)
            conn.commit()
            print(f"{time.asctime()}更新最新数据完毕")
        else:
            print(f"{time.asctime()}已经是最新数据")
    except:
        traceback.print_exc()
    finally:
        closeConn(conn, cursor)
```

```python
# 插入历史数据
def insertHistory():
    cursor = None
    conn = None
    try:
        dic = getTencentData()[0]    # 0 是历史数据字典， 1 是最新详细数据字典
        print(f"{time.asctime()}插入历史数据开始")
        conn, cursor = getConn()
        sql = "insert into history values(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
        for k,v in dic.items():
            #item格式 {'2020-04-22': {'confirm':4, 'suspect':0, 'heal':0, 'dead':1}}
            cursor.execute(sql, [k, v.get("confirm"), v.get("confirm_add"), 
                                 v.get("suspect"), v.get("suspect_add"),
                                 v.get("heal"), v.get("heal_add"),
                                 v.get("dead"), v.get("dead_add")])
        conn.commit()
        print(f"{time.asctime()}插入历史数据结束")
    except:
        traceback.print_exc()
    finally:
        closeConn(conn, cursor)
```

```python
# 更新历史数据
def updataHistory():
    cursor = None
    conn = None
    try:
        dic = getTencentData()[0]    # 0 是历史数据字典， 1 是最新详细数据字典
        print(f"{time.asctime()}历史数据更新开始")
        conn, cursor = getConn()
        sql = "insert into history values(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
        sql_query = "select confirm from history where ds=%s"
        for k,v in dic.items():
            #item格式 {'2020-04-22': {'confirm':4, 'suspect':0, 'heal':0, 'dead':1}}
            if not cursor.execute(sql_query, k):
                cursor.execute(sql, [k, v.get("confirm"), v.get("confirm_add"), 
                                     v.get("suspect"),v.get("suspect_add"),
                                     v.get("heal"), v.get("heal_add"),
                                     v.get("dead"), v.get("dead_add")])
        conn.commit()
        print(f"{time.asctime()}历史数据更新完毕")
    except:
        conn.rollback()
        traceback.print_exc()
    finally:
        closeConn(conn, cursor)
```

## 存储百度热搜数据

```python
import pymysql
def getConn():
    # 建立连接
    conn = pymysql.connect(host="127.0.0.1",
                           user = "root",
                           password = "",
                           db = "cov2019")
    # 创建游标
    cursor = conn.cursor()
    return conn,cursor

def closeConn(conn, cursor):
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```

```python
from selenium.webdriver import Chrome,ChromeOptions
import time
import traceback

def getBaiduHot():
    option = ChromeOptions()  #创建爱你谷歌浏览器实例
    option.add_argument("--headless")
    option.add_argument("--no-sandbox")

    url = 'https://voice.baidu.com/act/virussearch/virussearch?from=osari_map&tab=0&infomore=1'
    browser = Chrome(options=option,executable_path='./chromedriver.exe')
    browser.get(url)

    # 点击查看更多
    btn = browser.find_element_by_css_selector('#ptab-0 > div > div.VirusHot_1-5-6_32AY4F.VirusHot_1-5-6_2RnRvg > section > div')
    btn.click()
    time.sleep(1)

    # 获取热搜标签
    c = browser.find_elements_by_xpath('//*[@id="ptab-0"]/div/div[1]/section/a/div/span[2]')
    context = [i.text for i in c]   #获取标签内容
    print(context)
    return context

def updateHotsearch():
    cursor = None
    conn = None
    try:
        context = getBaiduHot()
        print(f"{time.asctime()}开始更新热搜数据")
        conn, cursor = getConn()
        sql = "insert into hotsearch(dt,content) values(%s,%s)"
        ts = time.strftime("%Y-%m-%d %X")
        for i in context:
            cursor.execute(sql, (ts, i))   #插入数据
        conn.commit()
        print(f"{time.asctime()}热搜数据更新完毕")
    except:
        traceback.print_exc()
    finally:
        closeConn(conn, cursor)
```
